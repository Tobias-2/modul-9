```python
import datetime as dt

def count_seniority(row):

    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019,9,27)
    else:
        end_date = row['DateofTermination']

    return (end_date - row['DateofHire'])/np.timedelta64(364,'D')
```

```python
import pandas as pd
import numpy as np
from datetime import datetime  # Import the datetime module

# Convert 'DOB' column to datetime format
df['DOB'] = pd.to_datetime(df['DOB'], format='%m/%d/%y')

# Define the function to calculate age
def count_age(row):
    # Define the fixed date as a datetime object
    target_date = datetime(2014, 9, 29)  # Proper datetime object
    return (target_date - row['DOB']) / np.timedelta64(364, 'D')  # Compute age in years

# Apply the function to the DataFrame
df['Age'] = df.apply(lambda row: count_age(row), axis=1)

df
```

```python
sns.histplot(df['Age'], kde=True)
plt.xlim(left=0) 
plt.show()
```
