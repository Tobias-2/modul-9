```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
genres = pd.read_csv('tmdb_genres.csv')
```

```python
movies = pd.read_csv('tmdb_movies.csv')
```

```python
genres
```

```python
genres = genres.rename(columns = {'Unnamed: 0':'genre_id'})
genres
```

```python
movies
```

```python
movies1 = movies.sort_values(by='vote_average',ascending=False)
```

```python
q3_vote_count = movies1['vote_count'].quantile(0.75)
```

```python
filtered_movies1 = movies1[movies1['vote_count'] > q3_vote_count]
```

```python
movies1 = top_10_movies = filtered_movies1.sort_values(by='vote_average', ascending=False).head(10)
```

```python
movies1
```

```python
movies['release_date'] = pd.to_datetime(movies['release_date'], errors='coerce')
```

```python
movies['release_year'] = movies['release_date'].dt.year
```

```python
filtered_movies = movies[(movies['release_year'] >= 2010) & (movies['release_year'] <= 2016)]
```

```python
filtered_movies
```

```python
avg_values = filtered_movies.groupby('release_year')[['revenue', 'budget']].mean().reset_index()
```

```python
avg_values
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.bar(avg_values['release_year'], avg_values['revenue'] / 1e9, color='blue', label='Revenue')
line = axes.plot(avg_values['release_year'], avg_values['budget'] / 1e9, color='red', linewidth=2, label='Budget')
axes.set_ylabel('Amount of money in bilions', color='black')
axes.set_xlabel('Year')
axes.tick_params(axis='y', labelcolor='black')
plt.title('Average Revenue and Budget (2010-2016)', fontsize=14)
axes.legend(loc=(1.05,0.9))
```

```python
fig
```

```python
genre_avg_runtime = movies_with_genres.groupby('genres')['runtime'].mean().sort_values(ascending=False)
```

```python
genre_avg_runtime
```

```python
longest_runtime_genre = genre_avg_runtime.idxmax()
```

```python
filtered_movies_with_genres = movies_with_genres[movies_with_genres['genres'] == longest_runtime_genre]
```

```python
plt.figure(figsize=(10, 6))
plt.hist(filtered_movies_with_genres['runtime'], bins=20, color='blue', edgecolor='black')
plt.title(f"Histogram of Runtimes for {longest_runtime_genre} Movies")
plt.xlabel("Runtime (minutes)")
plt.ylabel("Frequency")
plt.show()
```
